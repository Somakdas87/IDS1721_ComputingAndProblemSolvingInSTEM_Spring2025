# **Population growth model**

### **Introduction to Population Growth Models**

Population growth models describe how populations evolve over time.
A common approach is to represent the population as a vector, where
each element corresponds to a specific group, such as age or stage.
Changes in the population are governed by a transition matrix that
describes how individuals move between groups.

### **Applications:**

* Ecology: Modeling species population dynamics.

* Economics: Studying workforce age distribution.

* Epidemiology: Tracking the spread of diseases across regions.

# **Problem Statement**

In a closed ecosystem, there are four species of animals, each influencing the population growth of the others. The population dynamics of the species are governed by the equation:

$$P_{t+1}=A \cdot P_t$$

Where:

* $P_t$ is a column vector representing the populations of the four species
at time $t$.

* $A$ is a $4 \times 4$ transition matrix representing growth rates and interactions
between species.

* $P_{t+1}$ is the population vector at time $t+1$.

# **Initial State**
* Initial population vector:
$$P_0=\begin{pmatrix} 100\\ 200\\150\\ 50 \end{pmatrix}$$

* Transition matrix:
$$A = \begin{pmatrix} 1.1 & 0.02 &0.01 & -0.01\\
0.01&1.05&-0.02&0.02\\
0.03&-0.01&1.08&0.01\\
-0.01 & 0.02&0.01&1.02 \end{pmatrix}$$

### **Understanding the Transition Matrix**
* The transition matrix models the interactions and growth dynamics of four species in a closed ecosystem.

* Each element of the matrix $A$ represents a specific interaction or growth rate between the species.


* Diagonal element $A_{ii}$ represent the intrinsic growth rate of each species.

* Example:
$A_{22} = 1.05$ means species $2$ grows at $5\%$.

* Off-diagonal elements $A_{ij} (i \neq j)$ represent interactions between
species:

  * A positive value indicates that the species has a supportive or beneficial influence.

  * A negative value indicates that the species has a competitive or detrimental influence.

* Example:

  * $A_{14} = -0.01$: Species $4$ inhibits the growth of Species $1$.

  * $A_{31} =0.03$: Species $1$ supports the growth of Species $3$.

### **Understanding the Population vector**
$$P_0=\begin{pmatrix} p_1(t)\\ p_2(t)\\p_3(t)\\ p_4(t) \end{pmatrix} \hspace{10mm} P_0=\begin{pmatrix} 100\\ 200\\150\\ 50 \end{pmatrix}$$

* Structure of the Population Vector:

  * $p_1(t)$: Population of Species $1$ at time $t$.

  * $p_2(t)$: Population of Species $2$ at time $t.$

  * $p_3(t)$: Population of Species $3$ at time $t$.

  * $p_4(t)$: Population of Species $4$ at time $t$.


* For example, at the initial time $(t=0)$

  * Species $1$ starts with a population of $100$,

  * Species $2$ starts with a population of $200$,
  
  ... and so on ...

# **Interpretation**
At each time step:

* The intrinsic growth of each species
(determined by the diagonal elements of
$A$) contributes to its new population.

* The interactions between species (off-
diagonal elements of $A$) modify the
population growth or decline by
influence of other species.

* The first entry of $P_1$ is calculated as:

$$p_1(1)=(1.1)(100)+(0.02)(200)+(0.01)(150)
+(-0.01)(50)=112$$

* Similar calculations are performed for
the other species.

Recall:

$$P_{t+1}=A \cdot P_t \hspace{40mm} P_1 = A \cdot P_0$$
<br>
$$ A = \begin{pmatrix} 1.1 & 0.02 &0.01 & -0.01\\
0.01&1.05&-0.02&0.02\\
0.03&-0.01&1.08&0.01\\
-0.01 & 0.02&0.01&1.02 \end{pmatrix} \hspace{20mm} P_0=\begin{pmatrix} 100\\ 200\\150\\ 50 \end{pmatrix} $$



The matrix-vector product in `numpy` can be obtained by `numpy.dot()`.

```python
# Example
A = np.array([[1, 2, 3],
              [4, 5, 6]])    # 2x3 matrix

x = np.array([1, 0, -1])     # vector of size 3

b = np.dot(A, x)             # Result: 2x1 vector
print(b)                     # Output: [-2 -2]
```

Using `numpy` arrays for the transition matrix and the population vector their product is obtained in general as
$$P_{t+1} = \text{numpy.dot}(A,P_t)$$

That is,
$$P_{1} = \text{numpy.dot}(A,P_0)$$
$$P_{2} = \text{numpy.dot}(A,P_1)$$
$$P_{3} = \text{numpy.dot}(A,P_2)$$
... and so on ...

```python
P0 = np.array([100, 200, 150, 50])
print(P0)

A = np.array([
    [1.1 , 0.02, 0.01, -0.01],
    [0.01, 1.05, -0.02, 0.02],
    [0.03, -0.01, 1.08, 0.01],
    [-0.01, 0.02, 0.01, 1.02]
    ])
print(A)

T_max = 50

population_matrix = np.zeros((len(P0), T_max+1))
population_matrix[:,0] = P0

for t in range(T_max):
    population_matrix[:,t+1] = np.dot(A,population_matrix[:,t])

print("Population Matrix:")
r,c = np.shape(population_matrix)
for i in range(r):
    for j in range(c):
        print(f"{population_matrix[i,j]:10.2f}",end="")
    print()
```



# **Task 2: Visualize the results**
* Plot the population of each species over time.
* Use a different color for each species and add appropriate labels, legends, and a title.

```python
#@title { vertical-output: true }
t = np.arange(0,T_max+1)

for i in range(len(P0)):
    plt.plot(t,population_matrix[i,:],label=f"species {i+1}",marker = "o")

plt.title("Population dynamics of four species")
plt.xlabel("Time (t)")
plt.ylabel("Population")
plt.ylim([0,1200])
plt.grid()
plt.legend()

plt.show()
```

```python
P0 = np.array([500000,300000,200000])
print(P0)

A = np.array([
    [0.85, 0.10, 0.05],
    [0.05, 0.90, 0.05],
    [0.10, 0.15, 0.75]
    ])
print(A)

T_max = 100

population_matrix = np.zeros((len(P0), T_max+1))
population_matrix[:,0] = P0

for t in range(T_max):
    population_matrix[:,t+1] = np.dot(A,population_matrix[:,t])

print("Population Matrix:")
r,c = np.shape(population_matrix)
for i in range(r):
    for j in range(c):
        print(f"{population_matrix[i,j]:12.2f}",end="")
    print()
```

```python
#@title { vertical-output: true }
t = np.arange(0,T_max+1)

for i in range(len(P0)):
    plt.plot(t[60:],population_matrix[i,60:],label=f"regions {i+1}",marker = "o")

plt.title("Population dynamics for regions")
plt.xlabel("Time (t)")
plt.ylabel("Population")

plt.grid()
plt.legend()

plt.show()
```
