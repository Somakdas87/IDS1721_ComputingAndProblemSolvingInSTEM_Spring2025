# **Predator-Prey Models**

### **Overview**
The predator-prey model describes interactions between two species:

* Prey (e.g., rabbits): species that reproduce naturally and are consumed by predators.

* Predators (e.g., foxes): species that consume prey to survive and reproduce.

While classical predator-prey models like Lotka-Volterra use nonlinear differential equations, a matrix-based approach offers a linear, simplified view of population dynamics using:

$$x_{t+1} = A x_t$$

Where:
* $x_t = \begin{pmatrix} prey_t \\ predator_t \end{pmatrix}$ is the population vector at time $t$
* $A$ is a transition matrix capturing interactions
* The entries of $A$ determine how prey and predator populations affect each other and themselves

### **General Form of the Matrix**
$$A = \begin{pmatrix} a&b \\ c&d \end{pmatrix}$$
* $a$: impact of prey on itself (e.g., reproduction rate)
* $b$: impact of predators on prey (usually negative)
* $c$: impact of prey on predators (food source, usually positive)
* $d$: impact of predators on themselves (e.g., natural mortality)

### **Example**


```python
A = np.array([
    [1.02, -0.03],   # prey growth and loss due to predation
    [0.02,  0.98]    # predator growth from eating prey, and natural decay
])
```

# **Behavior Scenarios**

### **Stabilization (Convergence)**
The populations settle to a stable point over time.

$[$ Eigenvalues of $A$ have magnitude less than 1 $]$

Typical of balanced ecosystems with resource constraints.



```python
A = np.array([
    [0.9, -0.1],
    [0.1,  0.8]
])
```

**Result: Both populations decay and settle to equilibrium values.**

### **Oscillations**
Predator and prey populations rise and fall cyclically.

$[$Eigenvalues of $A$ are complex conjugates with magnitude near or equal to 1.$]$

Mimics real-life predator-prey cycles.

```python
A = np.array([
    [1.01, -0.03],
    [0.03,  0.97]
])
```

**Result: Cyclical population changes over time (up-down-up-down).**

### **Divergence (Explosion or Collapse)**
Population sizes grow or shrink uncontrollably.

$[$Eigenvalues greater than 1 in magnitude.$]$

Happens when predator or prey overpowers the system.


```python
A = np.array([
    [1.1, -0.01],
    [0.01, 1.05]
])
```

**Result: Prey and/or predator population grows too fast and becomes unrealistic.**

### **Extinction**
Populations fall to zero or very low numbers.

Usually happens when prey dies out too quickly, or predators cannot sustain themselves.



```python
A = np.array([
    [0.7, -0.2],
    [0.0,  0.5]
])
```

**Result: Predators quickly die out if prey disappears, leading to extinction.**


# **Key Takeaways**

* Matrix models offer a linear, discrete-time approximation of predator-prey interactions.

* Behavior depends heavily on the structure and eigenvalues of the matrix.

* While not as realistic as nonlinear models, they are useful for simple simulations.

# **Examples**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time steps
T = 1000
x = np.zeros((2, T))
x[:, 0] = [30, 15]  # Initial populations [prey, predator]

# Transition matrix tuned for oscillation
A = np.array([
    [1.02, -0.03],
    [0.02,  0.98]
])

# Simulate the system
for t in range(1, T):
    x[:, t] = A @ x[:, t - 1]

# Plot results
time = np.arange(T)
plt.plot(time, x[0], label="Prey", color='green')
plt.plot(time, x[1], label="Predator", color='red')
plt.title("Oscillating Predator-Prey Dynamics (Matrix Model)")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.grid(True)
plt.show()
```
