### <b> Evaluating Arithmetic Expressions</b>
Let us demonstrate how Python evaluates expressions using examples.

```python
2+7
```

```python
2**3
```

```python
15/4
```

```python
15//4
```

```python
19%5
```

Let us now try somewhat more complicated expressions now.

Python follows the PEMDAS (Parentheses, Exponents, Multiplication and Division, Addition and Subtraction) rule for operator precedence.

```python
2 + (3 * 4 / 2**2)**2
```

In this example, the operations are performed in the following order:
<ol type = "1">
<li> 2 ** 2 = 4 (exponentiation)</li>
<li> 3 * 4 = 12 (multiplication) </li>
<li> 12 / 4 = 3.0 (division) </li>
<li> 3.0 ** 2 = 9.0 (exponentiation) </li>
<li> 2 + 9.0 = 11.0 (addition)

This order of evaluation follows Python's operator precedence rules:

Parentheses are evaluated first
Exponentiation (**) has the highest precedence among operators
Multiplication and division have equal precedence and are evaluated left to right
Addition has the lowest precedence in this expression

### Example
Lets consider a simple formula
for calculating the compounded interest on a bank deposit:
$$A = P(1+(r/100))^n$$
where,
<li>P is the initial deposit (the principal),
<li>r is the yearly interest rate in percent,
<li>n is the number of years,
<li>A is the final amount.

The task is now to write a program that computes A for given values of
P, r and n.

For instance, if
$$P = 100,r = 5.0, n = 7$$
we can use Python as a calculator and use the following line to get the result.

```python
print(100*(1+5.0/100)**7)
```

However, a better way to do this is to use variables and mathematical expressions using the variales.


```python
P=100
r=5.0
n=7
A=(P*(1+5.0/100)**7)
print(A)
```

This way the formula can be reused in future for a new set of parameters and we do not have to write a code-line specific to the value every time.

```python
P=200
r=7.0
n=4
A=(P*(1+5.0/100)**7)
print(A)
```

Note, that we are having to copy the formula over to use it. Later in this course we will be learning different ways to automate the process of using a formula repeatedly.
Note that, we can also use a better output to screen using the following line.

```python
print("Final amount is",A)
```
