```python
def hanoi_step(n, tower1, tower2, tower3, count):
    """
    Shows the moves for solving Towers of Hanoi problem,
    along with the step number

    Args:
        n: Number of disks to start with
        tower1: Label for the first tower
        tower2: Label for the second tower
        tower3: Label for the third tower

    Returns:
        count: The number of steps
    """
    if n == 1:
        count += 1
        print(f"Step {count: 3d} : Move disk {n} from tower {tower1} to tower {tower2}")
        return count
    count = hanoi_step(n-1, tower1, tower3, tower2, count) + 1
    print(f"Step {count: 3d} : Move disk {n} from tower {tower1} to tower {tower2}")
    count = hanoi_step(n-1,tower3,tower2,tower1, count)
    return count
```

```python
c = hanoi_step(4, 'A', 'B', 'C',0)
```

```python
def count(n):
    if n == 1:
        print(n,1)
        return 1

    c = count(n-1)+1
    print(n, c)
    return c
```
