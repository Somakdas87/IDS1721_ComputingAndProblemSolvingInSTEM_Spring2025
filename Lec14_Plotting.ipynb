# **Chapter 6: Arrays and Plotting**

In Python, arrays can be created using the built-in array module. However, they are not as commonly used as lists or NumPy arrays.

## **NumPy Arrays**

NumPy (`numpy`) is a powerful library that provides multi-dimensional arrays (ndarray) with advanced mathematical and scientific computing capabilities. NumPy arrays are faster, more efficient, and support a wide range of operations compared to Python's built-in arrays.

Example of a NumPy array:



```python
import numpy as np

# Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Accessing elements
print(arr[0])  # Output: 1
```

```python
# Performing operations
arr = arr * 2  # Multiplies each element by 2
print(arr)  # Output: [2 4 6 8 10]
```

```python
# Broadcasting
arr = arr + 2
print(arr)
```

```python
# Creating multi-dimensional arrays
matrix = np.array([[1, 2, 3], [4, 5, 6]])

print(matrix.shape)  # Output: (2, 3)
print(matrix[0, 1])  # Accessing row 0, column 1;  Output: 2
```

```python
print(matrix[1,0])
```

### **1D Arrays in NumPy**

A 1D array in NumPy is a simple, one-dimensional collection of elements of the same data type. It is similar to a list in Python but provides more functionality and efficiency for numerical computations.

## **Creating a 1D NumPy Array**

A one-dimensional array can be created using `numpy.array()` or functions like `numpy.arange()`, `numpy.linspace()`, and `numpy.zeros()`.

### **Example: Creating a 1D Array**

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)  # Output: [1 2 3 4 5]
print(type(arr))
```

## **Common Functions to Create a 1D Array**

1. Using `np.arange(start, stop, step)`


```python
arr = np.arange(1, 16, 3)  # Creates an array from 1 to 9 with step 2
print(arr)  # Output: [1 3 5 7 9]
```

2. Using `np.linspace(start, stop, num_points)`


```python
arr = np.linspace(0, 10, 5)  # 5 evenly spaced points between 0 and 10
print(arr)  # Output: [ 0.   2.5  5.   7.5 10. ]
```

3. Using `np.zeros(n)` and `np.ones(n)`


```python
arr = np.zeros(5)  # Creates an array of 5 zeros
print(arr)  # Output: [0. 0. 0. 0. 0.]

arr = np.ones(5)  # Creates an array of 5 ones
print(arr)  # Output: [1. 1. 1. 1. 1.]

arr2 = np.zeros((5,3))
print(arr2)
```

## **Accessing Elements in a 1D Array**
Elements in a NumPy array can be accessed using indexing and slicing.


```python
arr = np.array([10, 20, 30, 40, 50])

# Accessing elements
print(arr[0])  # Output: 10 (First element)
print(arr[-1]) # Output: 50 (Last element)

# Slicing
print(arr[1:4])  # Output: [20 30 40]
print(arr[:3])   # Output: [10 20 30] (First three elements)

print(arr[-4:-1])
```

## **Mathematical Operations on 1D Arrays**

NumPy allows performing element-wise operations efficiently.

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr * 2)   # Output: [ 2  4  6  8 10] (Multiplication)
print(arr + 10)  # Output: [11 12 13 14 15] (Addition)
print(arr ** 2)  # Output: [ 1  4  9 16 25] (Exponentiation)
```

### **Useful Properties of 1D Arrays**

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr.shape)   # Output: (5,)  (1D array with 5 elements)
print(arr.size)    # Output: 5 (Total number of elements)
print(arr.dtype)   # Output: int64 (Data type of elements)
```

* 1D arrays in NumPy provide an efficient way to handle numerical data.
* They support mathematical operations, indexing, slicing, and broadcasting, making them highly useful in scientific computing, data analysis, and machine learning.

## **Storing Points on a Curve in a One-Dimensional NumPy Array**

In this example, we will generate and store points on a mathematical curve (such as a sine wave) using a one-dimensional NumPy array.


```python
# import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the range of x values
x_values = np.linspace(0, 2 * np.pi, 100)  # 100 points between 0 and 2Ï€

# Compute y values for a sine curve
y_values = np.sin(x_values)
```

```python
# Display the first few values
print("X values:", x_values[:5])  # First 5 x values
print("Y values:", y_values[:5])  # First 5 y values (sin of x)
```

```python
# Plot the curve
plt.plot(x_values, y_values)
```

```python
# Plot the curve
plt.plot(x_values, y_values, label="y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine Curve")
plt.legend()
plt.grid()
plt.show()
```

### Explanation
1. We use `np.linspace(0, 2*np.pi, 100)` to generate 100 evenly spaced points between $0$ and $2\pi$.

2. The corresponding y-values are computed using `np.sin(x_values)`, storing the results in a one-dimensional NumPy array.
3. We print a few values to verify the data.
4. Finally, we plot the curve using matplotlib.pyplot.

This approach can be used to store points on any mathematical curve (e.g., cosine, exponential, quadratic functions) in a one-dimensional array.

# **Example: Python program to plot multiple functions in the same graph**

This program will:

Plot $y=sin(x)$ and $y=cos(x)$ on the same graph.

* Use numpy to generate x-values and compute sine and cosine.
* Use matplotlib to visualize the functions.


```python
# import numpy as np
# import matplotlib.pyplot as plt

# Generate x values from -2Ï€ to 2Ï€
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Compute y values for sine and cosine
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sine and cosine
plt.figure(figsize=(10, 3))
plt.plot(x, y_sin, label=r"$y = \sin(x)$", color='blue')
plt.plot(x, y_cos, label="y = cos(x)", color='red', linestyle="dashed")

# Labels and title
plt.xlabel("x-axis (radians)")
plt.ylabel("y-axis")
plt.title("Sine and Cosine Functions")

# Add grid, legend, and axes
plt.axhline(0, color='green', linewidth=1)
plt.axvline(0, color='green', linewidth=2)
plt.axvline(np.pi, color='green', linewidth=2)
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()

# Show the plot
plt.show()
```

### Explanation
1. Generate x-values:
  * `np.linspace(-2*np.pi, 2*np.pi, 400)` creates values in the interval $[-2\pi , 2 \pi]$
2. Compute y-values:
  * `np.sin(x)` gives sine values.
  * `np.cos(x)` gives cosine values.
3. Plot the Graphs:
  * `plt.plot(x, y_sin, label=r"$y = \sin(x)$", color='blue')` plots sine.
  * `plt.plot(x, y_cos, label=r"$y = \cos(x)$", color='red', linestyle="dashed")` plots cosine as a dashed line.
3. Enhance the Graph:
  * Axes are centered using `axhline(0)` and `axvline(0)`.
  * Grid and legend are added.
4. Display the Plot: `plt.show()`

# **Exercise: Plotting a Mathematical Function and Its Derivative in Python**

## Problem Statement

### Design a Python program that includes:

1. A user-defined function that evaluates a given mathematical function
$f(x)$.

2. A second function that approximates the derivative of $f(x)$ using the central difference method.

3. A program that plots both the function and its derivative on the same graph for visualization.

### Instructions

1. Define a function `f(x)` that computes the value of a given mathematical function. Use $f(x)=x^3-4x^2+2x+1$ as an example.
2. Implement another function `df(x, h=0.001)` that approximates the derivative of $f(x)$ using the central difference formula:
$f'(x) \approx \frac{f(x+h)-f(x-h)}{2h}$

3. Generate a range of $x$-values using `numpy.linspace()`, evaluate $f(x)$ and its derivative $f'(x)$ for these values.
Use matplotlib to plot both the function
$f(x)$ and its derivative $f'(x)$ on the same graph.

4. Label the axes, add a legend, and display the graph.

```python
# import numpy as np
# import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return x**3 - 4*x**2 + 2*x + 1  # Example function

# Define the derivative function using central difference
def df(x, h=0.001):
    return (f(x + h) - f(x - h)) / (2 * h)

# Generate x values for plotting
x_values = np.linspace(-2, 5, 400)  # Define the range of x
y_values = f(x_values)  # Compute function values
dy_values = df(x_values)  # Compute derivative values

# Plot the function and its derivative
plt.figure(figsize=(4, 3))
plt.plot(x_values, y_values, label=r'$f(x) = x^3 - 4x^2 + 2x + 1$', color='b')
plt.plot(x_values, dy_values, label=r"$f '(x)$ (Derivative)", color='r', linestyle='dashed')

# Add labels and title
plt.xlabel("x")
plt.ylabel("Function values")
plt.title("Plot of f(x) and its derivative")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

$$ y  = f(x_0) + f'(x_0) ( x - x_0)$$

# **Exercise: Plot a Piecewise defined Function**

Consider the piecewise function
$$ f(x) =
\begin{cases}
      x^2 & x\leq 0 \\
      x+2 & 0\leq x\leq 2 \\
      \sin x & x \geq 2
   \end{cases}
$$

Write a Python program to plot this function.

```python
# import numpy as np
# import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    if x <= 0:
        return x**2
    elif x >= 0 and x <= 2:
        return x + 2
    else:
        return np.sin(x)


# Generate x values for plotting
x_values = np.linspace(-2, 4, 500)  # Define the range of x
y_values = [f(x) for x in x_values]  # Compute function values elementwise


# Plot the function and its derivative
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Piecewise function', color='b')


# Add labels and title
plt.xlabel("x")
plt.ylabel("Function values")
plt.title("Piecewise defined function")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
